```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('chinook.db')
cursor = conn.cursor()

# Get a list of table names
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
table_names = cursor.fetchall()

# Loop through the table names and retrieve data
for table in table_names:
    table_name = table[0]
    
    print(f"Table: {table_name}")
    
    # Get column names
    cursor.execute(f"PRAGMA table_info({table_name});")
    columns = [column[1] for column in cursor.fetchall()]
    print("Columns:", columns)
    
    # Retrieve a few rows from the table
    cursor.execute(f"SELECT * FROM {table_name} LIMIT 3;")
    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    print("\n")

# Close the connection
conn.close()
```

## What are the three most preferred genres?

```python
%%sql 
select  i.CustomerId,
            g.Name,
        il.Trackid,il.UnitPrice,il.Quantity,
        sum(il.UnitPrice*il.Quantity) as total_price
    
    
    
    
    
    
    
From Track t
Join Genre  g
on t.GenreID=g.GenreId
Join InvoiceLine il
on t.TrackId=il.TrackId
join Invoice i
on i.InvoiceId=il.InvoiceId
Group by g.Name 
Order by  total_price desc
limit 10
```

### Which country was the best buyer in 2009?

```python
%%sql
select c.Country,
        sum(il.UnitPrice*il.Quantity) as total_price
    
From Customer c
Join Invoice i
on i.CustomerId=c.CustomerId
Join InvoiceLine il
on il.InvoiceId=i.InvoiceId
WHERE i.InvoiceDate BETWEEN '2009-01-01' AND '2009-12-26'
Group by c.Country
order by 2 desc
limit 5
```

### Which was the most popular album ever sold on the playlist?

```python
%%sql
select a.title,
        sum(il.InvoiceId) as Total_buy
From Album a
Join Track t  on a.AlbumId=t.AlbumId
Join Invoiceline il on il.TrackId=t.TrackId
Group by a.title
order by Total_buy desc
limit 15
```
