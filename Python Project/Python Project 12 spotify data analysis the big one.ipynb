#### Converting duration from millisecond to second on Spotify Tracks Dataset¶

```python
tracks['duration']=tracks['duration_ms'].apply(lambda x: round(x/1000))
tracks.drop('duration_ms',inplace=True,axis=1)
tracks.duration.head()
```

In the code snippet you provided, a new column named 'duration' is added to the DataFrame `tracks` based on the 'duration_ms' column. The values in the 'duration' column are calculated by applying a lambda function to each value in the 'duration_ms' column. Here's a breakdown of what the code does:

1. `tracks['duration_ms']`: This selects the 'duration_ms' column from the DataFrame `tracks`.

2. `.apply(lambda x: round(x/1000))`: This applies a lambda function to each value in the 'duration_ms' column. The lambda function divides each value by 1000 and then rounds the result to the nearest whole number using the `round()` function.

3. `tracks['duration'] = ...`: This assigns the result of the lambda function to a new column named 'duration' in the DataFrame `tracks`. Each value in the 'duration' column represents the duration of a track in seconds (rounded to the nearest whole number) based on the corresponding value in the 'duration_ms' column.

Overall, this code snippet converts the duration of each track from milliseconds to seconds and stores the rounded result in a new 'duration' column in the `tracks` DataFrame.

#### Correlation Heatmap between Variable using Pearson correlation method

```python
cm=tracks.drop(['key','mode','explicit'],axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
map=sns.heatmap(cm,annot=True,fmt='.1g',vmin=-1,vmax=1,center=0,cmap='inferno',linewidths=1,linecolor='Black')
map.set_title('Correlation Heatmap Between Variable')
map.set_xticklabels(map.get_xticklabels(),rotation=90)
```

In the code snippet you provided, a correlation matrix is calculated for the DataFrame `tracks` after dropping the columns 'key', 'mode', and 'explicit'. Then, a correlation heatmap is created using the seaborn library (`sns`) with the help of matplotlib (`plt`). Here's an explanation of each line:

1. `cm = tracks.drop(['key', 'mode', 'explicit'], axis=1).corr(method='pearson')`: This line drops the specified columns ('key', 'mode', 'explicit') from the DataFrame `tracks` using the `drop()` method with the `axis=1` parameter (indicating columns). The remaining DataFrame is then used to calculate the correlation matrix using the `corr()` method with the `method='pearson'` parameter, which computes the Pearson correlation coefficients between all pairs of variables.

2. `plt.figure(figsize=(14, 6))`: This line creates a new figure with a specific size of 14 inches in width and 6 inches in height using `plt.figure()`.

3. `map = sns.heatmap(cm, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='black')`: This line creates the correlation heatmap using the seaborn `heatmap()` function. It takes the correlation matrix (`cm`) as the input and sets various parameters for annotation (`annot=True`), number formatting (`fmt='.1g'`), color scale limits (`vmin=-1` and `vmax=1` for the range -1 to 1), centering (`center=0`), colormap (`cmap='inferno'`), linewidths between cells (`linewidths=1`), and line color (`linecolor='black'`). The resulting heatmap is stored in the variable `map`.

4. `map.set_title('Correlation Heatmap Between Variable')`: This line sets the title of the heatmap to 'Correlation Heatmap Between Variable' using the `set_title()` method of the heatmap object.

5. `map.set_xticklabels(map.get_xticklabels(), rotation=90)`: This line rotates the x-axis tick labels by 90 degrees using the `set_xticklabels()` method, ensuring they are legible when displayed vertically.

Overall, this code generates a correlation heatmap visualizing the pairwise correlations between the remaining variables in the `tracks` DataFrame after dropping the specified columns. The heatmap provides insights into the relationships and strength of correlations among the variables, helping to identify patterns or dependencies.



In the code snippet `tracks.drop(['key', 'mode', 'explicit'], axis=1)`, the `axis=1` parameter is used to indicate that the specified columns should be dropped along the column axis.

In pandas, the `axis` parameter is used to specify the axis along which an operation should be performed. In the context of DataFrames, `axis=0` refers to the row axis, and `axis=1` refers to the column axis.

So, in the given code snippet, `axis=1` specifies that the columns `'key'`, `'mode'`, and `'explicit'` should be dropped from the `tracks` DataFrame along the column axis. The resulting DataFrame will have these columns removed.



In the code `corr(method='pearson')`, the `method='pearson'` parameter specifies the correlation method to be used in the `corr()` method.

The `corr()` method in pandas is used to calculate the correlation between columns in a DataFrame. By default, it computes the Pearson correlation coefficient, which measures the linear relationship between two variables. The Pearson correlation coefficient ranges from -1 to 1, where:

- A value of 1 indicates a perfect positive linear correlation.
- A value of -1 indicates a perfect negative (inverse) linear correlation.
- A value of 0 indicates no linear correlation between the variables.

The Pearson correlation coefficient is widely used to assess the strength and direction of the linear relationship between two variables.

In the provided code, `method='pearson'` explicitly specifies that the Pearson correlation coefficient should be calculated for all pairs of variables in the DataFrame. This is the most commonly used correlation method, but other methods like 'kendall' and 'spearman' are also available in pandas to calculate different types of correlations (e.g., rank-based correlations).

####  Creating new column in tracks dataset(Year,Release Date)

```python
tracks['dates']=tracks.index.get_level_values('release_date')
tracks.dates=pd.to_datetime(tracks.dates)
years=tracks.dates.dt.year
```

In the code snippet you provided, the DataFrame `tracks` is being modified by adding a new column called 'dates' based on the index values. Then, the 'dates' column is converted to datetime format using the `pd.to_datetime()` function. Finally, a new Series called 'years' is created by extracting the year component from the 'dates' column using the `.dt.year` accessor. Here's a breakdown of each line:

1. `tracks['dates'] = tracks.index.get_level_values('release_date')`: This line extracts the values from the 'release_date' level of the index and assigns them to a new column called 'dates' in the DataFrame `tracks`. The `get_level_values()` method is used to access the values at a specific level of a MultiIndex, assuming that the DataFrame has a MultiIndex with a level named 'release_date'.

2. `tracks.dates = pd.to_datetime(tracks.dates)`: This line converts the 'dates' column to datetime format using the `pd.to_datetime()` function. It ensures that the values in the 'dates' column are treated as proper datetime objects, which enables performing various datetime operations on them.

3. `years = tracks.dates.dt.year`: This line creates a new Series called 'years' by extracting the year component from the 'dates' column using the `.dt.year` accessor. The `.dt` accessor in pandas provides a set of datetime-specific methods, and `year` is one of them, which extracts the year component from a datetime object.

By performing these operations, you now have a new 'dates' column in the `tracks` DataFrame containing datetime values. Additionally, the 'years' Series is created, which extracts the year component from the 'dates' column. This allows you to work with the year information separately for analysis or further manipulation.

#### Change in Duration of Songs wrt Year

```python
total_dr=tracks.duration
fig_dims=(18,7)
fig, ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years,y=total_dr,ax=ax,errwidth=False).set(title='Years vs Distribution')
plt.xticks(rotation=90)
```

In the code snippet you provided, a bar plot is created using the seaborn library (`sns`) and the matplotlib.pyplot library (`plt`). The plot compares the distribution of the 'duration' variable against the 'years' variable. Here's an explanation of each line:

1. `total_dr = tracks.duration`: This line assigns the 'duration' column from the DataFrame `tracks` to the variable `total_dr`. It selects the values of the 'duration' column.

2. `fig_dims = (18, 7)`: This line defines the dimensions of the figure to be created. The figure dimensions are set to 18 inches in width and 7 inches in height, and these values are assigned to the tuple variable `fig_dims`.

3. `fig, ax = plt.subplots(figsize=fig_dims)`: This line creates a new figure and axes using the `plt.subplots()` function. The `figsize` parameter is set to `fig_dims`, which determines the size of the figure. The resulting figure and axes are assigned to the variables `fig` and `ax`, respectively.

4. `fig = sns.barplot(x=years, y=total_dr, ax=ax, errwidth=False).set(title='Years vs Distribution')`: This line creates a bar plot using the seaborn `barplot()` function. It takes the 'years' variable as the x-axis values, the 'total_dr' variable as the y-axis values, and the `ax` parameter to specify the axes to be used for the plot. The `errwidth=False` parameter is used to disable the display of error bars. Additionally, the `set()` method is used to set the title of the plot to 'Years vs Distribution'. The resulting plot is assigned to the variable `fig`.

5. `plt.xticks(rotation=90)`: This line rotates the x-axis tick labels by 90 degrees using the `xticks()` function from matplotlib.pyplot. It ensures that the x-axis tick labels are displayed vertically, making them more readable when there are many categories.

Overall, this code generates a bar plot comparing the distribution of the 'duration' variable against the 'years' variable. The x-axis represents the years, while the y-axis represents the 'duration' values. Each bar in the plot represents the distribution of 'duration' for a particular year. The plot provides a visual representation of how the duration of tracks is distributed across different years.
